```python
# Note: this notebook requires system Tk (the `tkinter` stdlib module).
# On macOS + Homebrew Python: `brew install python-tk@3.12`.

import snappy
import tkinter
import spherogram

unknot = spherogram.Link([[1, 2, 2, 1]])
knot3_1 = spherogram.Link('3_1')
knot4_1 = spherogram.Link('4_1')
knot5_1 = spherogram.Link('5_1')
knot5_2 = spherogram.Link('5_2')
knot6_1 = spherogram.Link('6_1')
knot8_19 = spherogram.Link('8_19')
knot10_124 = spherogram.Link('10_124')
knotsquare = knot3_1.connected_sum(knot3_1.mirror())
```

```python
knot_list = [unknot]

for K in knot_list:
    print(K.name)
    pd_code = K.PD_code()
    K = snappy.Link(pd_code)
    K.view()
    tkinter.mainloop()

    x = input('Continue? (y/n)')
    if x == 'n':
        break
```
